```python
x = list(range(2, 10, 2))

icarl = [94.23, 65.948, 46.528, 41.942]
icarl_std = [1.254750971, 1.427592379, 1.583136128, 1.481796207]

dynamic_er = [95.222, 63.22, 43.056, 39.472]
dynamic_er_std = [1.133631333, 1.757270611, 1.378959753, 2.398597507]

wa = [94.326, 58.59, 36.398, 30.65]
wa_std = [0.930204279, 2.089772715, 1.664097353, 0.57497826]

lucir = [95.13666667, 65.51666667, 46.85, 43.78]
lucir_std = [1.634299034, 1.217223617, 1.385965368, 1.551998711]

podnet = [94.92333333, 59.96333333, 43.34, 31.80333333]
podnet_std = [1.955232297, 0.9579318, 0.081853528, 1.175471537]

ours = [94.742, 70.128, 52.01, 50.688]
ours_std = [1.144866804, 2.070173906, 1.254731047, 3.059398634]

plt.figure(figsize=(3, 2), dpi=300) # default figsize=(6.4, 4.8), dpi=100

joint = plt.errorbar(x[-1], 70.37666667, fmt='x', color='black', linewidth=1, markersize=3)
ours_curve= plt.errorbar(x, ours,yerr=ours_std, fmt='.-', color='red', linewidth=1, markersize=3, elinewidth=1, capthick=1, capsize=1)
icar_curve = plt.errorbar(x, icarl, yerr=icarl_std, fmt='.-',color='dodgerblue', linewidth=1, markersize=3, elinewidth=1, capthick=1, capsize=1)
dynamic_er_curve = plt.errorbar(x, dynamic_er, yerr=dynamic_er_std, fmt='.-', color='green', linewidth=1, markersize=3, elinewidth=1, capthick=1, capsize=1)
wa_curve = plt.errorbar(x, wa, yerr=wa_std, fmt='.-', color='m', linewidth=1, markersize=3, elinewidth=1, capthick=1, capsize=1)
lucir_curve = plt.errorbar(x, lucir, yerr=lucir_std, fmt='.-', color='sienna', linewidth=1, markersize=3, elinewidth=1, capthick=1, capsize=1)
podnet_curve = plt.errorbar(x, podnet, yerr=podnet_std, fmt='.-', color='mediumblue', linewidth=1, markersize=3, elinewidth=1, capthick=1, capsize=1)

plt.legend(handles=[joint, ours_curve, lucir_curve,  icar_curve, dynamic_er_curve, podnet_curve, wa_curve],
    labels=['Upbound', 'Ours', 'UCIR',  'iCaRL', 'DynaER', 'PODNet', 'WA'], bbox_to_anchor=(0, 0),fontsize=5 ,loc=3, borderaxespad=0)

y_ticks = np.arange(20, 101, 10)
plt.xticks(x, fontsize=8)
plt.yticks(y_ticks, fontsize=8)

plt.grid(ls='--', c='gray', axis='y') # linewidth=2

# plt.xlim((49, 101))
plt.ylim((20, 101))

# plt.title('Base 2 Increment 2 (skin8 buffer80)')

plt.xlabel('Number of classses', fontsize=8)
plt.ylabel('MCR (%)', fontsize=8)

plt.savefig('./skin8_b0i2_buffer40.png', bbox_inches='tight')

plt.show()
```
